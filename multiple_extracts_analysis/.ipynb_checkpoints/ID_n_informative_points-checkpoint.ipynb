# Defining functions

```python
def present_in_array_index(new_sample, array, size = 11):
    present = False
    new_sample = np.reshape(np.array(new_sample), (1,size))
    for i in range(array.shape[0]):
        if np.array_equiv(array[i,:],new_sample):
            present = True
            break
    return(present, i)
```

```python
def get_data_from_row(i, y_mean_total = None, y_std_total = None):
    return(y_mean_total[i], y_std_total[i])
```

```python
def filling_array_from_data(already_sampled, X_total = None, y_mean_total = None, y_std_total = None):
    means = []
    errors = []
    for row in already_sampled:
        present, i = present_in_array_index(row, X_total)
        mean, error = get_data_from_row(i, y_mean_total = y_mean_total, y_std_total = y_std_total)
        means.append(mean)
        errors.append(error)
    means = np.reshape(means, (np.shape(already_sampled)[0], 1))
    errors = np.reshape(errors, (np.shape(already_sampled)[0], 1))
    completed_array = np.concatenate((already_sampled, means, errors), axis = 1)

    return(already_sampled, means, errors)
```

```python
def select_current_best_model(X, y, models_number = 10, verbose = False, 
                             MLP = True,
                             visu = False):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")

    trained_model_list = []
    # Training all models
    for i in range(models_number):
        #Â X_train, X_test, y_train, y_test, y_error_train, y_error_test = sklearn.model_selection.train_test_split(X, y, y_std, test_size=0.2) 
        X_train, y_train = X, y
        if MLP:
            other_MLP = MLPRegressor(hidden_layer_sizes  = (10, 100,100, 20), solver ="adam", max_iter=20000, 
                                      early_stopping = True, learning_rate = "adaptive")
            other_MLP.fit(X_train, y_train.flatten())    
            trained_model_list.append(other_MLP)

            big_MLP = MLPRegressor(hidden_layer_sizes  = (100,100, 20),solver ="adam", max_iter=20000, 
                                      early_stopping = True, learning_rate = "adaptive")
            big_MLP.fit(X_train, y_train.flatten())    
            trained_model_list.append(big_MLP)


            medium_MLP = MLPRegressor(hidden_layer_sizes  = (40, 10), solver ="adam", max_iter=20000, 
                                      early_stopping = True, learning_rate = "adaptive")
            medium_MLP.fit(X_train, y_train.flatten())    
            trained_model_list.append(medium_MLP)
  
            small_MLP = MLPRegressor(hidden_layer_sizes  = (10), solver ="adam", max_iter=20000, 
                                      early_stopping = True, learning_rate = "adaptive")
            small_MLP.fit(X_train, y_train.flatten())    
            trained_model_list.append(small_MLP)
        
    # Evaluating all 
    all_scores = []
    for i in range(len(trained_model_list)):
        selected_mdoel = trained_model_list[i]
        y_pred = selected_mdoel.predict(X)
        score = sklearn.metrics.r2_score(y, y_pred)
        all_scores.append(score)

    try:
        best_index = all_scores.index(max(all_scores))
        best_score = all_scores[best_index]
    except ValueError:
        best_index = 0
    if verbose:
        print(all_scores)
        print("Best index is {}".format(best_index))
        print("Best score is {}".format(best_score))
    best_model = trained_model_list[best_index]
    if visu:
        sample_weight_total = (1/(y_std*y_std))
        
        model = best_model
        y_pred = model.predict(X)
        score = sklearn.metrics.r2_score(y, y_pred)
        fig, ax = plt.subplots()
        ax.scatter(y, y_pred, edgecolors=(0, 0, 0))
        ax.errorbar(y, y_pred,xerr = y_std, ls='none')
        ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
        ax.set_xlabel('Measured')
        ax.set_title("Model prediction for model {}: {}".format(i, score))
        ax.set_ylabel('Predicted')
        plt.show()
    return(best_model, best_score)
```

```python
def obtain_split_score(X_train, X_test, y_train, y_test, 
                      ensemble_size = 25, models_number_repeat = 10):
    # Training ensemble models
    
    ensemble_models = []
    for n in range(ensemble_size):
        current_best, best_score = select_current_best_model(X_train, y_train,
                                                            models_number = models_number_repeat, verbose = False, 
                                                             MLP = True,
                                                             visu = False)
        ensemble_models.append(current_best)
    # evaluating the models
    all_predictions = None
    all_predictions_on_train = None
    for model in ensemble_models:
        y_pred = model.predict(X_test)
        y_pred_train = model.predict(X_train)
        answer_array_pred = y_pred.reshape(X_test.shape[0], -1)
        answer_array_pred_train = y_pred_train.reshape(X_train.shape[0], -1)
        if all_predictions is None:
            all_predictions = y_pred.reshape(X_test.shape[0], -1)
            all_predictions_on_train = y_pred_train.reshape(X_train.shape[0], -1)
        else:
            all_predictions =np.concatenate((all_predictions, y_pred.reshape(X_test.shape[0], -1)), axis = 1)
            all_predictions_on_train =np.concatenate((all_predictions_on_train, y_pred_train.reshape(X_train.shape[0], -1)), axis = 1)

    y_pred, y_pred_std = np.mean(all_predictions, axis = 1), np.std(all_predictions, axis = 1)
    y_pred_train, y_pred_std_train = np.mean(all_predictions_on_train, axis = 1), np.std(all_predictions_on_train, axis = 1)
    score = sklearn.metrics.r2_score(y_test, y_pred)
    y_pred_train = y_pred_train.reshape(y_pred_train.shape[0], -1)
    y_pred_std_train = y_pred_std_train.reshape(y_pred_std_train.shape[0], -1)
    y_pred = y_pred.reshape(y_pred.shape[0], -1)
    y_pred_std = y_pred_std.reshape(y_pred_std.shape[0], -1)
    
    return(score, y_pred_train, y_pred_std_train, y_pred, y_pred_std)
```

```python
def handle_results(score, 
                   X_train, X_test, y_train, y_test, y_std_train, y_std_test, 
                   y_pred_train, y_pred_std_train, y_pred, y_pred_std,
                   best_tests):
    if len(best_tests) < 5:
        best_tests.append({"score": score, "X": X_train, 
                          "X_test": X_test, "y_train": y_train, 
                           "y_test": y_test, "y_std_train": y_std_train, "y_std_test": y_std_test, 
                          "y_pred_train": y_pred_train, "y_pred_std_train": y_pred_std_train, 
                           "y_pred": y_pred, "y_pred_std": y_pred_std})
    else:
        minimal_element = best_tests[0]
        for element in best_tests:
            score_element = element["score"]
            if score_element <minimal_element["score"]:
                minimal_element = element
        if score > minimal_element["score"]:
            best_tests.remove(minimal_element)
            best_tests.append({"score": score, "X": X_train, 
                          "X_test": X_test, "y_train": y_train, 
                           "y_test": y_test, "y_std_train": y_std_train, "y_std_test": y_std_test, 
                          "y_pred_train": y_pred_train, "y_pred_std_train": y_pred_std_train, 
                           "y_pred": y_pred, "y_pred_std": y_pred_std})
    return(best_tests)
```

```python
def handle_multiple_results(score_dict, X_train, best_tests):
    if len(best_tests) < 5:
        new_best_score = score_dict
        new_best_score["X"] = X_train
        best_tests.append(new_best_score)
    else:
        minimal_element = best_tests[0]
        for element in best_tests:
            score_element, X_element = element["mean_score"], element["X"]
            if score_element <minimal_element["mean_score"]:
                minimal_element = element
        if score_dict["mean_score"] > minimal_element["mean_score"]:
            best_tests.remove(minimal_element)
            new_best_score = score_dict
            new_best_score["X"] = X_train
            best_tests.append(new_best_score)
    return(best_tests)
```

```python
def save_results(best_tests, initial_max, base_name = "final", folder_to_save = "test", visu = False):
    index = 0
    for element in best_tests:
        index = index + 1
        score_element = element["mean_score"]
        headers = "nad;folinic_acid;coa;nucleo_mix;spermidin;pga;aa;trna;mg_gluta;camp;K_gluta"
        file_to_write = "{}/{}_{}.csv".format(folder_to_save, base_name, index)
        X = copy.deepcopy(element["X"])
        for i in range(X.shape[1]):
            X[:,i] = X[:,i]*initial_max[i]
        np.savetxt(file_to_write, X, delimiter=";", header = headers,fmt='%.6f')
        
```

```python
def obtaining_rest_of_data(already_sampled, X_total = None, y_mean_total = None, y_std_total = None):
    means = []
    errors = []
    X_test = None
    for row in X_total:
        present_in_starting, i = present_in_array_index(row, already_sampled)
        if not present_in_starting:
            present, i = present_in_array_index(row, X_total)
            mean, error = get_data_from_row(i, y_mean_total = y_mean_total, y_std_total = y_std_total)
            means.append(mean)
            errors.append(error)
            if X_test is None:
                X_test = np.reshape(row, (1, 11))
            else:
                X_test = np.concatenate((X_test, np.reshape(row, (1, 11))), axis = 0)
    means = np.reshape(means, (len(means), 1))
    errors = np.reshape(errors, (len(means), 1))
    completed_array = np.concatenate((X_test, means, errors), axis = 1)

    return(X_test, means, errors)
```

```python
def get_score_from_name(initial_max, base_folder ="search_informative", 
                        name_of_plate = "final_1", 
                        folder_to_save = "search_informative",
                        X_PS_total = X_data_PS, y_PS_total = y_data_PS, y_std_PS_total = y_std_data_PS, 
                        X_ORI_total = X_data_ORI, y_ORI_total = y_data_ORI, y_std_ORI_total = y_std_data_ORI,
                        X_AB_total = X_data_AB, y_AB_total = y_data_AB, y_std_AB_total = y_std_data_AB,
                        visu = True):
    
    file_to_read = "{}/{}.csv".format(base_folder, name_of_plate)
    my_current_data = np.genfromtxt(file_to_read, delimiter=';', skip_header  = 1, dtype = "float")
    X_tr_PS, y_tr_PS_d, y_std_tr_PS_d = filling_array_from_data(my_current_data,
                                               X_total = X_PS_total, 
                                                y_mean_total = y_PS_total, 
                                                y_std_total = y_std_PS_total)
    X_tr_ORI, y_tr_ORI_d, y_std_tr_ORI_d = filling_array_from_data(my_current_data,
                                                          X_total = X_ORI_total, 
                                                          y_mean_total = y_ORI_total, 
                                                           y_std_total = y_std_ORI_total)
    X_tr_AB, y_tr_AB_d, y_std_tr_AB_d = filling_array_from_data(my_current_data,
                                                            X_total = X_AB_total, 
                                                            y_mean_total = y_AB_total, 
                                                            y_std_total = y_std_AB_total)
    po = np.allclose(X_tr_PS, X_tr_ORI)
    pa = np.allclose(X_tr_PS, X_tr_AB)
    assert po and pa
    
    X_test_PS, y_test_PS_d, y_std_test_PS_d = obtaining_rest_of_data(my_current_data,
                                                        X_total = X_PS_total, 
                                                        y_mean_total = y_PS_total, 
                                                        y_std_total = y_std_PS_total)
    X_test_ORI, y_test_ORI_d, y_std_test_ORI_d = obtaining_rest_of_data(my_current_data,
                                                        X_total = X_ORI_total, 
                                                        y_mean_total = y_ORI_total, 
                                                        y_std_total = y_std_ORI_total)
    X_test_AB, y_test_AB_d, y_std_test_AB_d = obtaining_rest_of_data(my_current_data,
                                                        X_total = X_AB_total, 
                                                        y_mean_total = y_AB_total, 
                                                        y_std_total = y_std_AB_total)
    
    for i in range(X_tr_PS.shape[1]):
        X_tr_PS[:,i] = X_tr_PS[:,i]/max(X_tr_PS[:,i])    
    for i in range(X_tr_ORI.shape[1]):
        X_tr_ORI[:,i] = X_tr_ORI[:,i]/max(X_tr_ORI[:,i])    
    for i in range(X_tr_AB.shape[1]):
        X_tr_AB[:,i] = X_tr_AB[:,i]/max(X_tr_AB[:,i])    
    for i in range(X_test_PS.shape[1]):
        X_test_PS[:,i] = X_test_PS[:,i]/max(X_test_PS[:,i])    
    for i in range(X_test_ORI.shape[1]):
        X_test_ORI[:,i] = X_test_ORI[:,i]/max(X_test_ORI[:,i])    
    for i in range(X_test_AB.shape[1]):
        X_test_AB[:,i] = X_test_AB[:,i]/max(X_test_AB[:,i])

    po = np.allclose(X_test_PS, X_test_ORI)
    pa = np.allclose(X_test_PS, X_test_AB)
    assert po and pa
    
    score_ORI, y_ORI_tr, y_ORI_std_tr, y_ORI_pr, y_ORI_std_pr = obtain_split_score(X_tr_ORI, X_test_ORI, y_tr_ORI_d, y_test_ORI_d)
    score_PS, y_PS_tr, y_PS_std_tr, y_PS_pr, y_PS_std_pr = obtain_split_score(X_tr_PS, X_test_PS, y_tr_PS_d, y_test_PS_d)
    score_AB, y_AB_tr, y_AB_std_tr, y_AB_pr, y_AB_std_pr = obtain_split_score(X_tr_AB, X_test_AB, y_tr_AB_d, y_test_AB_d)
    
    mean_score = np.mean([score_ORI, score_PS, score_AB])

    for i in range(X_tr_PS.shape[1]):
        X_tr_PS[:,i] = X_tr_PS[:,i]*initial_max[i]
    for i in range(X_test_PS.shape[1]):
        X_test_PS[:,i] = X_test_PS[:,i]*initial_max[i]
    for i in range(X_tr_ORI.shape[1]):
        X_tr_ORI[:,i] = X_tr_ORI[:,i]*initial_max[i]
    for i in range(X_test_ORI.shape[1]):
        X_test_ORI[:,i] = X_test_ORI[:,i]*initial_max[i]
    for i in range(X_tr_AB.shape[1]):
        X_tr_AB[:,i] = X_tr_AB[:,i]*initial_max[i]
    for i in range(X_test_AB.shape[1]):
        X_test_AB[:,i] = X_test_AB[:,i]*initial_max[i]
    
    total_PS_tr = np.concatenate((X_tr_PS, y_tr_PS_d, y_std_tr_PS_d, y_PS_tr, y_PS_std_tr), axis = 1)
    total_PS_pr = np.concatenate((X_test_PS, y_test_PS_d, y_std_test_PS_d, y_PS_pr, y_PS_std_pr), axis = 1)
    
    total_ORI_tr = np.concatenate((X_tr_ORI, y_tr_ORI_d, y_std_tr_ORI_d, y_ORI_tr, y_ORI_std_tr), axis = 1)
    total_ORI_pr = np.concatenate((X_test_ORI, y_test_ORI_d, y_std_test_ORI_d, y_ORI_pr, y_ORI_std_pr), axis = 1)
    
    total_AB_tr = np.concatenate((X_tr_AB, y_tr_AB_d, y_std_tr_AB_d, y_AB_tr, y_AB_std_tr), axis = 1)
    total_AB_pr = np.concatenate((X_test_AB, y_test_AB_d, y_std_test_AB_d, y_AB_pr, y_AB_std_pr), axis = 1)
    
    # Training data:
    headers = "nad;folinic_acid;coa;nucleo_mix;spermidin;pga;aa;trna;mg_gluta;camp;K_gluta;y_data;y_data_std;y_pr;y_std_pr"
    file_to_write_PS = "{}/{}_PS_train.csv".format(folder_to_save, name_of_plate)
    np.savetxt(file_to_write_PS, total_PS_tr, delimiter=";", header = headers,fmt='%.6f')
    file_to_write_ORI = "{}/{}_ORI_train.csv".format(folder_to_save, name_of_plate)
    np.savetxt(file_to_write_ORI, total_ORI_tr, delimiter=";", header = headers,fmt='%.6f')
    file_to_write_AB = "{}/{}_AB_train.csv".format(folder_to_save, name_of_plate)
    np.savetxt(file_to_write_AB, total_AB_tr, delimiter=";", header = headers,fmt='%.6f')
    # Testing data
    file_to_write_PS = "{}/{}_PS_test.csv".format(folder_to_save, name_of_plate)
    np.savetxt(file_to_write_PS, total_PS_pr, delimiter=";", header = headers,fmt='%.6f')
    file_to_write_ORI = "{}/{}_ORI_test.csv".format(folder_to_save, name_of_plate)
    np.savetxt(file_to_write_ORI, total_ORI_pr, delimiter=";", header = headers,fmt='%.6f')
    file_to_write_AB = "{}/{}_AB_test.csv".format(folder_to_save, name_of_plate)
    np.savetxt(file_to_write_AB, total_AB_pr, delimiter=";", header = headers,fmt='%.6f')
    
    dict_scores = {}
    dict_scores["mean_score"] = mean_score
    dict_scores["score_ORI"] = score_ORI
    dict_scores["score_PS"] = score_PS
    dict_scores["score_AB"] = score_AB
    
    if visu:
        #Â PS
        fig, ax = plt.subplots()
        ax.scatter(y_test_PS_d, y_PS_pr, edgecolors=(0, 0, 0))
        ax.errorbar(y_test_PS_d, y_PS_pr, 
                            xerr = y_std_test_PS_d, yerr = y_PS_std_pr, ls='none')
        ax.set_xlabel('Measured')
        ax.set_title("R squared is: {} for {}".format(round(score_PS, 2), name_of_plate))
        ax.set_ylabel('Predicted')
        ax.scatter(y_tr_PS_d, y_PS_tr, edgecolors='red')
        ax.errorbar(y_tr_PS_d, y_PS_tr, 
                            xerr = y_std_tr_PS_d, yerr = y_PS_std_tr, 
                    ls='none', ecolor = "red")
        ax.plot([y_test_PS_d.min(), y_test_PS_d.max()],
                 [y_test_PS_d.min(), y_test_PS_d.max()], 'k--', lw=4)
        name_for_plotting = "{}/{}_PS_model_vs_data.png".format(folder_to_save, name_of_plate)
        plt.savefig(name_for_plotting)
        plt.show()
        # AB
        fig, ax = plt.subplots()
        ax.scatter(y_test_AB_d, y_AB_pr, edgecolors=(0, 0, 0))
        ax.errorbar(y_test_AB_d, y_AB_pr, 
                            xerr = y_std_test_AB_d, yerr = y_AB_std_pr, ls='none')
        ax.set_xlabel('Measured')
        ax.set_title("R squared is: {} for {}".format(round(score_AB, 2), name_of_plate))
        ax.set_ylabel('Predicted')
        ax.scatter(y_tr_AB_d, y_AB_tr, edgecolors='red')
        ax.errorbar(y_tr_AB_d, y_AB_tr, 
                            xerr = y_std_tr_AB_d, yerr = y_AB_std_tr, 
                    ls='none', ecolor = "red")
        ax.plot([y_test_AB_d.min(), y_test_AB_d.max()],
                 [y_test_AB_d.min(), y_test_AB_d.max()], 'k--', lw=4)
        name_for_plotting = "{}/{}_AB_model_vs_data.png".format(folder_to_save, name_of_plate)
        plt.savefig(name_for_plotting)
        plt.show()
        # ORI
        fig, ax = plt.subplots()
        ax.scatter(y_test_ORI_d, y_ORI_pr, edgecolors=(0, 0, 0))
        ax.errorbar(y_test_ORI_d, y_ORI_pr, 
                            xerr = y_std_test_ORI_d, yerr = y_ORI_std_pr, ls='none')
        ax.set_xlabel('Measured')
        ax.set_title("R squared is: {} for {}".format(round(score_ORI, 2), name_of_plate))
        ax.set_ylabel('Predicted')
        ax.scatter(y_tr_ORI_d, y_ORI_tr, edgecolors='red')
        ax.errorbar(y_tr_ORI_d, y_ORI_tr, 
                            xerr = y_std_tr_ORI_d, yerr = y_ORI_std_tr, 
                    ls='none', ecolor = "red")
        ax.plot([y_test_ORI_d.min(), y_test_ORI_d.max()],
                 [y_test_ORI_d.min(), y_test_ORI_d.max()], 'k--', lw=4)
        name_for_plotting = "{}/{}_ORI_model_vs_data.png".format(folder_to_save, name_of_plate)
        plt.savefig(name_for_plotting)
        plt.show()
        
    return(dict_scores)

```

```python
n_list = [8, 16, 20, 24, 32, 64]

split_values_dict = {20: 0.80, 16:0.84, 8:0.92, 24:0.76, 32:0.68, 64: 0.37}


for n in n_list:
    best_tests = []
    all_scores = []

    folder_to_save = "search_informative/informative_{}".format(n)

    for test_number in range(nbr_combin):
        X_tr, X_test, y_tr_ORI, y_test_ORI, y_tr_PS, y_test_PS, y_tr_AB, y_test_AB = train_test_split(X_data, y_data_ORI, y_data_PS, y_data_AB, test_size=split_values_dict[n])
        score_ORI, a,b,c,d = obtain_split_score(X_tr, X_test, y_tr_ORI, y_test_ORI,
                                               ensemble_size = ensemble_size, models_number_repeat = models_number_repeat)
        score_PS, a,b,c,d = obtain_split_score(X_tr, X_test, y_tr_PS, y_test_PS,
                                                ensemble_size = ensemble_size, models_number_repeat = models_number_repeat)
        score_AB, a,b,c,d = obtain_split_score(X_tr, X_test, y_tr_AB, y_test_AB,
                                                  ensemble_size = ensemble_size, models_number_repeat = models_number_repeat)
        mean_score = np.mean([score_ORI, score_PS, score_AB])
        dict_scores = {}
        dict_scores["mean_score"] = mean_score
        dict_scores["score_ORI"] = score_ORI
        dict_scores["score_PS"] = score_PS
        dict_scores["score_AB"] = score_AB
        all_scores.append(mean_score)
        best_tests = handle_multiple_results(dict_scores, X_tr, best_tests)
        best_tests_scores = [element["mean_score"] for element in best_tests]
        if test_number % nbr_to_save == 0:
            save_results(best_tests, initial_max, base_name = "iteration_{}".format(test_number), folder_to_save = folder_to_save, visu = True)

    save_results(best_tests, initial_max, base_name = "final".format(test_number), folder_to_save = folder_to_save)
    dict_of_scores = {}
    for file_name in ["final_1", "final_2", "final_3", "final_4", "final_5"]:
        print("Currently processing file {} for {} points".format(file_name, n))
        short_name = file_name
        score = get_score_from_name(initial_max = initial_max,
                                    name_of_plate = short_name, 
                                    base_folder = folder_to_save, 
                                    folder_to_save = folder_to_save, 
                                    visu = visu)
        dict_of_scores[short_name] = score
    with open("{}/dict_of_scores_{}.json".format(folder_to_save, n), "w") as json_handler:
        json.dump(dict_of_scores, json_handler, indent = 2)
```
